### Indexing

```python
import numpy as np 
```

```python
arr = np.array([10, 20, 30, 40, 50])
```

```python
print(arr[0])   # 10
print(arr[-1])  # 50
```

```python
# [20,30,40]
arr[1:4] 
```

```python
# [10,20,30]
# arr[0:3] -> gives same result as arr [:3]
arr[:3]
```

```python
# arr [10,30,50]
arr[::2]
# arr[0:5:2] -> gives same result as arr[::2]
```
